```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('wine.csv')
dataset
```

```python
dataset.info()
```

```python
data2=dataset.iloc[:,1:].values
data2
```

```python
# normalise as we are using K-Means later
from sklearn.preprocessing import scale
data_norm=scale(data2)
data_norm
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=13)
data_pca=pca.fit_transform(data_norm)
data_pca
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
cum_sum=np.cumsum(np.round(pca.explained_variance_ratio_,4)*100)
cum_sum
```

```python
# merging data with first 3 components 
data_merge=pd.concat([dataset['Type'],pd.DataFrame(data_pca[:,:3],columns=['PC1','PC2','PC3'])],axis=1)
data_merge
```

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(20,20))
dendrogram = sch.dendrogram(sch.linkage(data_norm, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
hc
```

```python
y_hc = hc.fit_predict(data_norm)
y_hc
```

```python
data_hier=dataset.copy()
data_hier['Cluster_Id']=y_hc
data_hier
```

```python
data_hier['Cluster_Id'].value_counts()
```

## K-Means Cluster

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 5):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 5), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(data_norm)
```

```python
y_kmeans
```

```python
data_kmean=dataset.copy()
data_kmean['Cluster_Id']=y_kmeans
data_kmean
```

```python
data_kmean['Cluster_Id'].value_counts()
```
